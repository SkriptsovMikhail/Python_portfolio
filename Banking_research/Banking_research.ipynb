```python
# Функция для преобразования названия столбца в camelCase
def to_camel_case(s):
    parts = s.split()
    # Объединяем первую часть с остальными, капитализируя их
    return parts[0].lower() + ''.join(word.capitalize() for word in parts[1:])

# Применяем функцию к названиям столбцов
new_columns = [to_camel_case(col) for col in df.columns]
df.columns = new_columns
```

```python
# Удаление дубликатов
df.drop_duplicates(inplace=True)

# Удаление пропусков
df.dropna(inplace=True)
```

Построить диаграмму ящик с усами по параметру Сумма транзакций не представляется возможным


```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

# Преобразование столбца "customerdob" в формат даты
df['customerdob'] = pd.to_datetime(df['customerdob'])

# Получение текущей даты
today = datetime.today()

# Вычисление возраста каждого клиента
df['age'] = (today - df['customerdob']).dt.days / 365.25

# Фильтрация по возрасту (удаление строк с возрастом 0 или менее 18 лет)
df = df[(df['age'] > 18) & (df['age'] != 0)]

# Анализ возраста клиентов
print("Статистические данные о возрасте клиентов:")
print(df['age'].describe())

# Группировка по возрастным категориям (например, по десятилетиям)
df['age_group'] = pd.cut(df['age'], bins=[18, 28, 38, 48, 58, 68], labels=['18-28', '28-38', '38-48', '48-58', '58-68'])
print("\nРаспределение клиентов по возрастным группам:")
print(df['age_group'].value_counts())

# Построение гистограммы распределения возраста
plt.hist(df['age'], bins=10, range=(18, 68)) # Устанавливаем диапазон для гистограммы
plt.xlabel('Возраст')
plt.ylabel('Количество клиентов')
plt.title('Распределение возраста клиентов')
plt.show()

# Построение ящика с усами по возрасту
plt.figure(figsize=(8, 6))  # Увеличиваем размер графика
plt.boxplot(df['age'], vert=False, patch_artist=True, showmeans=True, showfliers=False)  # Убираем выбросы (showfliers=False)

# Получение значений Q1, Q3 и Медианы
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
median = df['age'].median()

# Отображение значений Q1, Q3 и Медианы на графике
plt.text(Q1, 0.8, 'Q1: {:.1f}'.format(Q1), ha='center', va='top')
plt.text(Q3, 0.7, 'Q3: {:.1f}'.format(Q3), ha='center', va='top')
plt.text(median, 0.6, 'Медиана: {:.1f}'.format(median), ha='center', va='top')

plt.xlabel('Возраст')
plt.title('Ящик с усами по возрасту клиентов')
plt.show()
```

Проанализируем возраст клиентов на момент совершения первой транзакции


```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

# Преобразование столбцов дат в формат даты
df['customerdob'] = pd.to_datetime(df['customerdob'])
df['transactiondate'] = pd.to_datetime(df['transactiondate'])

# Сортировка датафрейма по customerid и transactiondate
df = df.sort_values(by=['customerid', 'transactiondate'])

# Создание нового датафрейма с первой транзакцией для каждого клиента
first_transactions = df.groupby('customerid').first().reset_index()

# Вычисление возраста на момент первой транзакции
first_transactions['age_at_first_transaction'] = (first_transactions['transactiondate'] - first_transactions['customerdob']).dt.days / 365.25

# Фильтрация по возрасту (удаление строк с возрастом 0 или менее 18 лет)
first_transactions = first_transactions[(first_transactions['age_at_first_transaction'] > 18) & (first_transactions['age_at_first_transaction'] != 0)]

# Анализ возраста клиентов на момент первой транзакции
print("Статистические данные о возрасте клиентов на момент первой транзакции:")
print(first_transactions['age_at_first_transaction'].describe())

# Группировка по возрастным категориям (например, по десятилетиям)
first_transactions['age_group'] = pd.cut(first_transactions['age_at_first_transaction'], bins=[18, 28, 38, 48, 58, 68, 78, 88, 98, 108], labels=['18-28', '28-38', '38-48', '48-58', '58-68', '68-78', '78-88', '88-98', '98-108'])
print("\nРаспределение клиентов по возрастным группам на момент первой транзакции:")
print(first_transactions['age_group'].value_counts())

# Построение гистограммы распределения возраста на момент первой транзакции
plt.figure(figsize=(8, 6))
plt.hist(first_transactions['age_at_first_transaction'], bins=10)
plt.xlabel('Возраст на момент первой транзакции')
plt.ylabel('Количество клиентов')
plt.title('Распределение возраста клиентов на момент первой транзакции')
plt.show()

# Построение ящика с усами по возрасту на момент первой транзакции
plt.figure(figsize=(8, 6))  # Увеличиваем размер графика
plt.boxplot(first_transactions['age_at_first_transaction'], vert=False, patch_artist=True, showmeans=True, showfliers=False)  # Убираем выбросы (showfliers=False)

# Получение значений Q1, Q3 и Медианы
Q1 = first_transactions['age_at_first_transaction'].quantile(0.25)
Q3 = first_transactions['age_at_first_transaction'].quantile(0.75)
median = first_transactions['age_at_first_transaction'].median()

# Отображение значений Q1, Q3 и Медианы на графике
plt.text(Q1, 0.8, 'Q1: {:.1f}'.format(Q1), ha='center', va='top')
plt.text(Q3, 0.7, 'Q3: {:.1f}'.format(Q3), ha='center', va='top')
plt.text(median, 0.6, 'Медиана: {:.1f}'.format(median), ha='center', va='top')

plt.xlabel('Возраст на момент первой транзакции')
plt.title('Ящик с усами по возрасту клиентов на момент первой транзакции')
plt.show()
```

Выводы

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates  # Импортируем mdates

# Сгруппируйте транзакции по датам
transactions_by_date = df.groupby('transactiondate').agg(
    count_transactions=('transactionid', 'count'),
    total_amount=('transactionamount_(inr)', 'sum')
)

# Преобразуйте индекс в формат datetime
transactions_by_date.index = pd.to_datetime(transactions_by_date.index)

# Постройте линейный график количества транзакций
plt.figure(figsize=(10, 6))
plt.plot(transactions_by_date.index, transactions_by_date['count_transactions'])
plt.xlabel('Дата')
plt.ylabel('Количество транзакций')
plt.title('График количества транзакций по дням')

# Добавьте подписи месяцев на оси X
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Используем mdates.MonthLocator


plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Постройте линейный график суммы транзакций
plt.figure(figsize=(10, 6))
plt.plot(transactions_by_date.index, transactions_by_date['total_amount'])
plt.xlabel('Дата')
plt.ylabel('Сумма транзакций')
plt.title('График суммы транзакций по дням')

# Добавьте подписи месяцев на оси X
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Используем mdates.MonthLocator


plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Вывод:
1. Как видно из графиков, с середины августа по конец сентября наблюдались провалы в количествве транзакций, в первой половине октября активность восстанавливалась, но к середине октября снова снижалась
2. То же самое можно сказать и по суммам транзакций, что связано со снижением количества транзакций

```python
import pandas as pd
import matplotlib.pyplot as plt

# Сгруппируйте транзакции по датам и посчитайте среднюю сумму транзакции
transactions_by_date = df.groupby('transactiondate').agg(
    avg_transaction_amount=('transactionamount_(inr)', 'mean')
)

# Преобразуйте индекс в формат datetime
transactions_by_date.index = pd.to_datetime(transactions_by_date.index)

# Постройте линейный график средней суммы транзакции
plt.figure(figsize=(10, 6))
plt.plot(transactions_by_date.index, transactions_by_date['avg_transaction_amount'])
plt.xlabel('Дата')
plt.ylabel('Средняя сумма транзакции')
plt.title('График средней суммы транзакции по дням')

# Добавьте подписи месяцев на оси X
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Сгруппируйте транзакции по датам и просуммируйте сумму транзакций
sales_by_date = df.groupby('transactiondate')['transactionamount_(inr)'].sum()

# Преобразуйте индекс в формат datetime
sales_by_date.index = pd.to_datetime(sales_by_date.index)

# Постройте линейный график динамики продаж
plt.figure(figsize=(10, 6))
plt.plot(sales_by_date.index, sales_by_date)
plt.xlabel('Дата')
plt.ylabel('Сумма продаж')
plt.title('Динамика продаж по дням')

# Добавьте подписи месяцев на оси X
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Вывод:
1. Как видно из графика, с середины августа по конец сентября наблюдались провалы в продажах, в первой половине октября активность восстанавливалась, но к середине октября снова снижалась
2. Это также связано с празднованием в Индии религиозных праздников

```python
import pandas as pd
import matplotlib.pyplot as plt

# Создайте возрастные группы с шагом 10 лет, начиная с 18
df['age_group'] = pd.cut(df['age'], bins=range(18, 100, 10), labels=False, right=False)

# Сгруппируйте продажи по возрастным группам и просуммируйте их
sales_by_age_group = df.groupby('age_group')['transactionamount_(inr)'].sum()

# Отсортируйте продажи по убыванию
sales_by_age_group = sales_by_age_group.sort_values(ascending=False)

# Найдите возрастную группу с максимальными продажами
most_profitable_age_group = sales_by_age_group.idxmax()

# Вычислите долю продаж этой группы от общего объема продаж
total_sales = sales_by_age_group.sum()
profitability_share = sales_by_age_group[most_profitable_age_group] / total_sales

# Выведите результат
print(f"Доля продаж самой платежеспособной возрастной группы ({most_profitable_age_group}) составляет: {profitability_share:.2%}")

# Постройте столбчатую диаграмму
plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_age_group.index, sales_by_age_group.values)

# Добавьте подписи значений к столбцам
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom')

# Настройте график
plt.xlabel('Возрастная группа')
plt.ylabel('Сумма продаж')
plt.title('Сумма продаж по возрастным группам клиентов')
plt.xticks(sales_by_age_group.index, [f'{i}-{i+10}' for i in range(18, 58, 10)])  # Измените диапазон
plt.grid(False)  # Уберите сетку
plt.show()
```

Вывод:
1. Самая платежеспособная группа клиентов в возрасте 18-28 лет
2. Доля продаж в этой возрастной группе 43%

```python
import pandas as pd
import matplotlib.pyplot as plt

# Преобразование миллисекунд в минуты
df['transaction_duration_minutes'] = df['transactiontime'] / (1000 * 60)

# Построение гистограммы
plt.figure(figsize=(10, 6))  # Установите желаемый размер фигуры
plt.hist(df['transaction_duration_minutes'], bins=20, edgecolor='black')  # Измените bins для изменения количества столбцов
plt.xlabel('Длительность транзакции (миллисекунды)')
plt.ylabel('Количество транзакций')
plt.title('Распределение длительности транзакций')

# Убираем сетку
plt.grid(False)

# Добавляем подписи значений
for i, v in enumerate(plt.hist(df['transaction_duration_minutes'], bins=20)[0]):
    plt.text(plt.hist(df['transaction_duration_minutes'], bins=20)[1][i] + (plt.hist(df['transaction_duration_minutes'], bins=20)[1][1] - plt.hist(df['transaction_duration_minutes'], bins=20)[1][0])/2,
             v + 1, str(int(v)), ha='center', va='bottom')

plt.show()

# Построение ящика с усами
plt.figure(figsize=(8, 6))  # Установите желаемый размер фигуры
plt.boxplot(df['transaction_duration_minutes'], vert=True, patch_artist=True, showmeans=True)  # vert=True - вертикальный ящик с усами

# Получение значений для подписей
median = df['transaction_duration_minutes'].median()
q1 = df['transaction_duration_minutes'].quantile(0.25)
q3 = df['transaction_duration_minutes'].quantile(0.75)

# Добавление подписей
plt.text(1.1, median, f"Медиана: {median:.2f}", ha='left', va='center')
plt.text(1.1, q1, f"Q1: {q1:.2f}", ha='left', va='center')
plt.text(1.1, q3, f"Q3: {q3:.2f}", ha='left', va='center')

plt.xlabel('Длительность транзакции (минуты)')
plt.title('Распределение длительности транзакций')

plt.show()

```
